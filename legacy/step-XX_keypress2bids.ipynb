This notebook will fetch the keypress data from the experiment and convert the coherent and incoherent responses into BIDS events format.

```python
import os

# Set the path to the directory containing the keypress files
keypress_dir = os.path.join('..', 'data', 'keypress-raw')

# Set the path to the directory where the BIDS-formatted keypress files will be saved
keypress_bids_dir = os.path.join('..', 'data', 'keypress-bids')
```
