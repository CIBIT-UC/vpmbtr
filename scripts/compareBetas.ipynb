```python
# Settings
data_dir = '/DATAPOOL/VPMB/BIDS-VPMB-SPE'
space_label = "MNI152NLin2009cAsym"
derivatives_folder = "derivatives/fmriprep23/fmriprep"
nilearn_folder = "derivatives/nilearn_glm/group"
#smoothing_fwhm = 6.0
#high_pass_hz = 0.003
```

```python
task_label_list = ['AA','UA']
acq_label_list = ['0500','0750','1000','2500']
```

```python
# Import ROI values to a dataframe
import pandas as pd
import numpy as np
import os
import glob


# Create a dataframe with all the ROI values
df = pd.DataFrame(columns=['task','acq','beta'])

# Loop over tasks and acquisitions
for tt in task_label_list:
    for aa in acq_label_list:

        task_label = task_label_list[tt]
        acq_label = acq_label_list[aa]

        # Get the beta values
        beta_file = os.path.join(data_dir,nilearn_folder,
                                 f"group_task-{task_label}_acq-{acq_label}_contrast-ambiguousMinusStatic_c-bonferroni_p-0.05_clusterk-50_mean-beta-hMT-per-subject.tsv")
        
        beta_df = pd.read_csv(beta_file,sep='\t')
        beta_df['task'] = task_label
        beta_df['acq'] = acq_label
        beta_df['beta'] = beta_df['beta'].astype(float)

        # Append to the dataframe
        df = df.append(beta_df,ignore_index=True)




    




```
